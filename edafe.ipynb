```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading the dataset

df = pd.read_csv('weight-height.csv')

df_male = df[0:5000]
df_female = df[5000:]
df_male.drop("Gender", axis=1)
df_female.drop("Gender", axis=1)
```

```python
df.head()
```

```python
plt.scatter(x=df_male["Height"], y=df_male["Weight"], c="blue")
plt.xlabel("Height")
plt.ylabel("weight")
plt.title("Male height vs weight")
```

```python
plt.scatter(x=df_female["Height"], y=df_female["Weight"], c="pink")
plt.xlabel("Height")
plt.ylabel("weight")
plt.title("female height vs weight")
```

```python
df_male.to_csv("weight-height-male.csv")
df_female.to_csv("weight-height-female.csv")
```
